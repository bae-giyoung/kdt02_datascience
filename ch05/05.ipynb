#### 축을 바꾸는 것, 데이터를 붙이는 것! => 더 알아보기!

```python
df_state = pd.read_csv('../datas/state-abbrevs.csv')
df_state.head(), df_state.shape, df_state.info()
```

```python
df_area = pd.read_csv('../datas/state-areas.csv')
df_area.head(), df_area.shape, df_area.info()
```

```python
df_pop = pd.read_csv('../datas/state-population.csv')
df_pop.head(), df_pop.shape, df_pop.info()
```

```python
# pandas에서의 join은 merge로 !!!!!
# SELECT S.state, S. , A.area (sq.mi), P.population
# FROM STATE S, AREA A, POP P
# WHERE S.state = A.state
# AND S.abbreviation = P.state/region
# how='inner'가 default 사실 적을 필요 없음. sql에서 WHERE S.state = A.state, column이름이 같은 경우! 
# 다른 경우 left_on='state', right_on='state'
pd.merge(df_state, df_area, how='inner', left_on='state', right_on='state') 
```

#### P.169 ~ P.172
##### 병합을 JOIN처럼 생각하자!

```python
# how='inner'가 default. 사실 적을 필요 없음. sql에서 WHERE S.state = A.state, column이름이 같은 경우! 
# 다른 경우 left_on='state', right_on='state'
m_df = pd.merge(df_state, df_area, how='inner', left_on='state', right_on='state') # inner join
```

```python
m_df.head()
```

```python
# SELECT S.state, S. , A.area (sq.mi), P.population
# FROM STATE S, AREA A, POP P
# WHERE S.state = A.state
# AND S.abbreviation = P.state/region
pd.merge(df_pop, m_df, left_on='state/region', right_on='abbreviation') # 2448 : inner join은 데이터 있는 것만 나오니 적게 나옴
```

```python
# SELECT S.state, S. , A.area (sq.mi), P.population
# FROM P LEFT OUTER JOIN STATE.S ON S.abbreviation = P.state/region
#   AND STATE S INNER JOIN AREA A ON S.state = A.state
pd.merge(df_pop, m_df, how='left', left_on='state/region', right_on='abbreviation') # 2544
```
