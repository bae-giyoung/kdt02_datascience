# iris 독립변인 4개 -> 4*4 -> 16개 그리기 (i,i)는 자기 자신과의 관계

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
df = pd.DataFrame(X, columns=feature_names)
df['species'] = [target_names[i] for i in y]
target_names
```

```python
feature_names_kr = ['꽃받침 길이', '꽃받침 너비', '꽃입 길이', '꽃입 너비']
target_names_kr = ['세토사', '버시컬러', '버지니카']
```

```python
colors = ["#FF6B6B", "#4ECDC4", "#45B7D1"]
species_colors = {target_names[i]: colors[i] for i in range(3)}
```

### 특성 간 관계 분석

###### 히스토그램 표로 되어 있는 도수 분포를 정보 그림으로 나타낸 것
###### 도수 분포는

```python
fig, axes = plt.subplots(4, 4, figsize=(16,16)) # nrows, ncols
fig.suptitle("Iris 데이터 특성 간 관계 분석", fontsize=50) # 제목, 라벨 무조건 달아야 한다!
# (1,3)짜리 subplots을 가장 많이 씀 -> 처음에 내가 예측한 데이터, 머신러닝으로 향상된 데이터, 딥러닝 으로 향상된 데이터

for i in range(4):
    for j in range(4):
        #print(i, j)
        ax = axes[i, j] # axes는 축 배열이다
        if i == j:
            # ===== 막대 그래프, 히스토!!!!! -> 도수 분포를 위한 데이터 정리
            for k, species in enumerate(target_names): # enumerate는 인덱스를 사용하고 싶을 때!
                data = df[df['species'] == species].iloc[:, i]
                ax.hist(data, 
                        bins=12,  # 구간 개수 지정 bins의 기본값 10, bins가 너무 차이 나서 => 이런 것 조정하는 기준과 방법에 대해!!!!
                        color=colors[k], 
                        alpha=0.3, 
                        #density=True, # 적분값을 1로 만들어서 -> 목적 알아보기
                        label=target_names_kr[k])
                ax.set_title(feature_names_kr[i])
                ax.set_ylabel("갯수") # density 사용하면 "밀도"로 바꿔야 함
                if i == j:
                    ax.legend()
        else:
            # ===== 산점도!!!!!!
            for k, species in enumerate(target_names):
                data = df[df['species'] == species]
                x = data.iloc[:, j]
                y = data.iloc[:, i] # y가 종속변인
                ax.scatter(x, 
                           y, 
                           color=colors[k], 
                           alpha=0.6, 
                           label=target_names_kr[k],
                           s=35,
                           edgecolors='black')
            ax.set_xlabel(feature_names_kr[i])
            ax.set_ylabel(feature_names_kr[j])

            corr = df.iloc[:, [j,i]].corr().iloc[0,1]
            ax.text(
                0.05, # x 좌표에 놔라
                0.95, # y 좌표에 놔라
                f"r={corr:2f}",
                transform=ax.transAxes # 그래프 안에 넣어라, 축 변경
            )
        ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
