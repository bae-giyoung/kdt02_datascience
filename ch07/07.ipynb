```python
# 
plt.plot(x, np.sin(x-0), linestyle="solid")
plt.plot(x, np.sin(x-1), linestyle="dashed")
plt.plot(x, np.sin(x-2), linestyle="dashdot")
plt.plot(x, np.sin(x-3), linestyle="dotted")
plt.show()
```

```python
plt.plot(x, np.sin(x-0), "-") # solid
plt.plot(x, np.sin(x-1), "--") # dashed
plt.plot(x, np.sin(x-2), "-.") # dashdot
plt.plot(x, np.sin(x-3), ":") # dotted
plt.show()
```

```python
plt.plot(x, np.sin(x-0), "-g") # solid green
plt.plot(x, np.sin(x-1), "--c") # dashed cyan
plt.plot(x, np.sin(x-2), "-.k") # dashdot black
plt.plot(x, np.sin(x-3), ":r") # dotted red
plt.show()
```

```python
iris = load_iris()
iris # DataFrame이 아니라 Python의 Hash map
```

```python
iris.keys()
```

```python
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
df = pd.DataFrame(X)
df
```

```python
df = pd.DataFrame(X, columns=feature_names)
df
```

```python
t = []
for i in y:
    t.append(str(iris.target_names[i]))
t
```

```python
df = pd.DataFrame(X, columns=feature_names)
df["species"] = [str(target_names[i]) for i in y] # y = iris.target
df.head()
```

```python
df.tail()
```

```python
df.corr(numeric_only=True) # 상관 관계 => 이걸 통계학에서 뭐라고 하는지 확인! # 숫자만 보여줄래?
```

```python
means = df.groupby("species").mean()

colors = ["red", "blue", "green"]
species_list = iris.target_names
```

```python
means.head()
```

```python
plt.figure(figsize=(12, 8))
# plt.plot("sepal length (cm)", means.loc["setosa"][0]) # 안되는 것에 대해서 설명하기
for i , species in enumerate(species_list):
    plt.plot(iris.feature_names, means.loc[species], color=colors[i], label=species)

# 무조건 label, title 반드시 붙여야 함!
plt.title("종별 특성 평균값 비교")
plt.xlabel("특성")
plt.ylabel("평균값 (cm)")
plt.legend()
plt.show()
```
